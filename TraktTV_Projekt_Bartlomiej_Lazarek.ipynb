```python
import requests

client_id = "YOUR_CLIENT_ID"
access_token = "YOUR_ACCESS_TOKEN"

headers = {
    'Content-Type': 'application/json',
    'trakt-api-version': '2',
    'trakt-api-key': client_id,
    'Authorization': f'Bearer {access_token}'
}
```

## GET 1 – popularny film

```python
res_popular = requests.get('https://api.trakt.tv/movies/popular', headers=headers)
movies = res_popular.json()
movie_id = movies[0]['ids']['trakt']
print(f"Popularny film: {movies[0]['title']} ({movies[0]['year']}) – ID: {movie_id}")
```

## GET 2 – Pobierz listę popularnych filmów


```python
response = requests.get('https://api.trakt.tv/movies/popular', headers=headers)
popular_movies = response.json()
for movie in popular_movies[:5]:
    print(f"{movie['title']} ({movie['year']})")
```

## GET 3 – szczegóły jednego filmu

```python
res_info = requests.get(f'https://api.trakt.tv/movies/{movie_id}', headers=headers)
print(res_info.json())
```

## POST – dodaj film do watchlisty

```python
url_post = 'https://api.trakt.tv/sync/watchlist'
data = {"movies": [{"ids": {"trakt": movie_id}}]}
res_post = requests.post(url_post, headers=headers, json=data)
print(res_post.status_code, res_post.json())
```

## PATCH 1 – Oznacz film jako 'do obejrzenia później' z notatką

```python
data_patch_extra = {
  "movies": [
    {
      "ids": {
        "trakt": movie_id
      },
      "notes": "Zobacz w weekend"
    }
  ]
}
url_patch_extra = 'https://api.trakt.tv/sync/watchlist'
res_patch_extra = requests.post(url_patch_extra, headers=headers, json=data_patch_extra)
print(res_patch_extra.status_code, res_patch_extra.json())
```

## PUT (PATCH) 2 – Oznacz film jako obejrzany

```python
data_patch = {
  "movies": [
    {
      "ids": {
        "trakt": 1
      }
    }
  ]
}
url_patch = 'https://api.trakt.tv/sync/history'
res_patch = requests.post(url_patch, headers=headers, json=data_patch)
print(res_patch.status_code, res_patch.json())
```

## DELETE – usunięcie z watchlisty

```python
headers = {
    "Content-Type": "application/json",
    "trakt-api-version": "2",
    "trakt-api-key": client_id,
    "Authorization": f"Bearer {access_token}"
}

res_delete = requests.delete("https://api.trakt.tv/checkin", headers=headers)

print("DELETE status:", res_delete.status_code)
print("DELETE response:", res_delete.text)
```
